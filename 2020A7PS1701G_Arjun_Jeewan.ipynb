```python
#importing relevant libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as stats
from scipy.stats import beta
```

```python
# Plotting a prior probability distribution over theta when y0= 42 and y1= 18
b0, b1 = 43, 19 #b0= y0+1, b1= y1+1
prior = beta(b1, b0) #make beta distribution with parameters b0 and b1
x= np.linspace(0, 1, 250) #make 250 equally spaced points between 0 and 1
plt.xlabel('prior',fontsize=20) #label the axes
plt.ylabel(r"$\theta$", fontsize=20) #label the axes
plt.grid() #add a grid
plt.plot (x, prior.pdf(x)) #plot the prior distribution
```

```python
def plot_posterior(n1, n2, m1, m2, ax=None):
    x = np.linspace(0, 1, 250) #make 250 equally spaced points between 0 and 1
    y = stats.beta.pdf(x, n1+m1, n2+m2) #make m2 distribution with parameters n1+m1 and n2+m2
        
    if not ax:
        fig, ax = plt.subplots()

    ax.plot(x, y)
    ax.set_xlabel(r"$\theta$", fontsize=20)
    ax.set_ylabel(r"$P(\theta|D)$", fontsize=20)
    ax.set_title("Posterior after {} heads, {} tails, \
                 Prior: BetaPDF({},{})".format(n1, n2, m1, m2))
    #show grid lines
    ax.grid(True)
    return ax

                
plot_posterior(6, 9, 18, 42) #plotting the posterior distribution with observed data 6 heads and 9 tails
```

Visually, MAP probability appears to be around 0.31, since our priors are very confident, they dictate the probability more than the observed trials. 
Exact value of MAP probability is calculated by the formula found in question 2(B)
(6+18-1)/[(6+18-1)+(9+42-1)]= 23/73 = 0.31

MLE is calculated by the formula found in question 2(A)
6/(6+9)= 0.4


```python
#Here, since the priors are extremely confident, posterior should be very close to the prior.
#creating prior distribution with y0=420 and y1=180 and plotting posterior distribution with observed data 6 heads and 9 tails
plot_posterior(6, 9, 180, 420)
#since the priors are extremely confident, the posterior is almost the same as the prior and the posterior has very little dependence on data
```

```python
#Here since the priors are not confident, posterior should be very different from the prior.
#creating prior distribution with y0=32 and y1=28 and plotting posterior distribution with observed data 6 heads and 9 tails
plot_posterior(6, 9, 28, 32)
# Since the priors are less confident, the posterior is dictated by the data more.
```
